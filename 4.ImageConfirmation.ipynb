```python
import os
import webbrowser
```

```python
def decdeg2dms(dd):
    mult = -1 if dd < 0 else 1
    mnt,sec = divmod(abs(dd)*3600, 60)
    deg,mnt = divmod(mnt, 60)
    return mult*deg, mult*mnt, mult*sec
```

```python
# Function to construct the Google Maps URL
def construct_google_maps_url(latitude, longitude):
    lat_deg, lat_min, lat_sec = decdeg2dms(abs(latitude))
    lat_deg = int(lat_deg)
    lat_min = int(lat_min)

    lon_deg, lon_min, lon_sec = decdeg2dms(abs(longitude))
    lon_deg = int(lon_deg)
    lon_min = int(lon_min)
    
    lat_dir = 'N' if latitude >= 0 else 'S'
    lon_dir = 'E' if longitude >= 0 else 'W'
    url = f"https://www.google.com/maps/place/{lat_deg}°{lat_min}'{lat_sec:.1f}\"{lat_dir}+{lon_deg}°{lon_min}'{lon_sec:.1f}\"{lon_dir}"
    return url
```

```python
# Folder path containing the files
folder_path = "C:/Users/Rojano/Desktop/AlivevsDeadKudzu/output/Alive/top20"
folder = os.listdir(folder_path)
```

```python
# Dictionary to store the results
results = {}
```

```python

# Open the text file containing the coordinates already saved
with open("C:/Users/Rojano/Desktop/AlivevsDeadKudzu/results.txt", "r") as file:
    # Iterate through each line in the file
    for line in file:
        # Split the line by ":" to separate the filename and kudzu presence
        parts = line.strip().split(":")
        
        # Extract the filename and kudzu presence
        filename = parts[0].strip()
        kudzu_presence = parts[1].strip()
        
        # Add the filename and kudzu presence to the dictionary
        results[filename] = kudzu_presence
```

```python
print(f"{len(results)} images analyzed out of {len(folder)}.")
```

```python
analyzed_count = 0

# Iterate over files in the folder
for filename in folder:
    # Skip directories
    if os.path.isdir(os.path.join(folder_path, filename)):
        continue
        
    # Check if the file has already been analyzed
    if filename in results:
         print(f"File {filename} has already been analyzed.")
         continue
        
    # Get coordinates from the filename (assuming filename is in the format "image_latitude_longitude.jpg")
    # Split the filename by underscores
    parts = filename.split("_")
    
    # Extract latitude and longitude from the filename
    latitude = float(parts[1])
    longitude = float(parts[2])
    
    # Additional information (if present)
    additional_info = "_".join(parts[3:])
    
    # Print the extracted values for verification
    print("\nFilename:", filename)
    print("Latitude:", latitude)
    print("Longitude:", longitude)
    print("Additional Info:", additional_info)
    
    # Construct Google Maps URL
    google_maps_url = construct_google_maps_url(latitude, longitude)
    print(google_maps_url)
    
    # Open Google Maps in the browser
    webbrowser.open(google_maps_url)
    
    # Ask for input about Kudzu presence
    kudzu_presence = input(f"Do you see Kudzu at {latitude}, {longitude}? (yes/no/dead): ")
    
     # Check if the user wants to quit
    if kudzu_presence.lower() == 'q':
        break
    
    # Increment the count of images analyzed
    analyzed_count += 1
    
    # Save the result in the dictionary
    results[filename] = kudzu_presence

# Print the total count of images analyzed
print(f"Total images analyzed: {analyzed_count}")
```

```python
# Save the results to a file
with open("C:/Users/Rojano/Desktop/AlivevsDeadKudzu/results.txt", "w") as f:
    for filename, kudzu_presence in results.items():
        f.write(f"{filename} : {kudzu_presence}\n")
```

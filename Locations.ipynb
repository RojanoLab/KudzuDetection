```python
import pandas as pd
file_name = 'revisits.csv'
df = pd.read_csv(file_name)
```

```python
# Filter rows where 'Status' is 'Positive'
rev_positive = df[df['Status'] == 'Positive']
```

```python
rev_positive
```

```python
rev_positive = rev_positive[['Latitude','Longitude']]
rev_positive = rev_positive.dropna(how='all')
```

```python
rev_positive
```

```python
import pandas as pd

file_name = 'mappings.csv'
df = pd.read_csv(file_name, encoding='ISO-8859-1')  # or use encoding='latin1'
```

```python
df_positive = df[df['Status'] == 'Positive']
```

```python
df_positive = df_positive[['Latitude', 'Longitude']]
df_positive = df_positive.dropna(how='all')
```

```python
df_positive
```

```python
combined_df = pd.concat([rev_positive, df_positive], ignore_index=True)
```

```python
combined_df = combined_df.drop_duplicates()
```

```python
combined_df
```

```python
combined_df.to_csv('cleaned_data.csv', index=False)
```
